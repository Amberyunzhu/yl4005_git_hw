```python
import sys
from gurobipy import *

def lp_optimize(rows, cols, c, obj_sense, A, sense, rhs, lb, ub, vtype,
                   solution):

    model = Model()

  # Add variables to model
    for j in range(cols):
        model.addVar(lb=lb[j], ub=ub[j], vtype=vtype[j])
    model.update()
    vars = model.getVars()

  # Populate A matrix
    for i in range(rows):
        expr = LinExpr()
        for j in range(cols):
            if A[i][j] != 0:
                expr += A[i][j]*vars[j]
        model.addConstr(expr, sense[i], rhs[i])

  # Populate objective
    obj = QuadExpr()
#  for i in range(cols):
#    for j in range(cols):
#      if Q[i][j] != 0:
#        obj += Q[i][j]*vars[i]*vars[j]
    for j in range(cols):
        if c[j] != 0:
            obj += c[j]*vars[j]
    model.setObjective(obj, obj_sense)

  # Write model to a file
    model.update()
    model.write('autoGenerated.lp')

  # Solve
    model.optimize()

    if model.status == GRB.Status.OPTIMAL:
        x = model.getAttr('x', vars)
        for i in range(cols):
            solution[i] = x[i]
        return True
    else:
        return False
```

```python
import sys
from gurobipy import *
rows = len(nodes)
cols = len(edges)
x = [[0 for j in range(len(edges))] for i in range(4039)]
y = [0 for i in range(4039)]
y[0] = 1
for item in enumerate(edges):
    count = item[0]
    outflow = item[1][0]
    inflow = item[1][1]
    x[outflow][count] = 1
    x[inflow][count] = -1   
```

```python
timelist = list()

for node in nodes:
    exposure_time = 0
    if node == 0:
        timelist.append(exposure_time)
    else:
        y[node] = -1
        obj_sense = GRB.MINIMIZE
        A = x
        c = [1 for i in range(cols)]
        sense = [GRB.EQUAL for i in range(rows)]
        rhs = y
        lb = [0]*(cols)
        ub = [GRB.INFINITY]*(cols)
        vtype = [GRB.CONTINUOUS]*(cols)
        sol = [0]*(cols)
        success = lp_optimize(rows, cols, c, obj_sense, A, sense, rhs, lb, ub, vtype, sol)
        exposure_time = sum(sol)
        timelist.append(exposure_time)
        y[node] = 0
    print('node',node)

average=sum(timelist)/len(timelist)
```
